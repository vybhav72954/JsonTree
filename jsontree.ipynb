```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_excel("datafile.xlsx", index_col=False)
df
```

```python
# We have the DateFrame ready
# Now we can select the Columns as per requirements
```

```python
df1 = df.drop(['#',	"DESIGNATION", "DEPARTMENT"], axis=1)
df1
```

```python
df1.rename(columns = {'MANAGER EMPLOYEE_ID':'MANAGER_EMPLOYEE_ID'}, inplace = True)
```

```python
df1
```

```python
merged_df1 = df1.merge(df1, left_on="MANAGER_EMPLOYEE_ID", right_on="EMPLOYEE_ID", how='left')
```

```python
merged_df1
```

```python
manager_df = df1.merge(df1[["EMPLOYEE_ID", "NAME"]], left_on="MANAGER_EMPLOYEE_ID", right_on="EMPLOYEE_ID", how='left').rename({"NAME_y": "MANAGER_NAME"}).drop("EMPLOYEE_ID_y", axis=1)
```

```python
names = {"NAME_y": "MANAGER_NAME", 
        "NAME_x": "NAME", 
        "EMPLOYEE_ID_x": "EMPLOYEE_ID"}
manager_df.rename(columns=names, inplace=True)
```

```python
manager_df
```

```python
# Now after the Join, we have the manager Name.
```

```python
# Now we need to set up the hierarchy and convert it JSON (or dict)
# Using Lambda and GroupBy
```

```python
tree = (manager_df.groupby(['MANAGER_EMPLOYEE_ID','MANAGER_NAME'])['EMPLOYEE_ID','NAME']
       .apply(lambda x: x.to_dict('r'))
       .reset_index(name='data')
       .groupby('MANAGER_EMPLOYEE_ID')['MANAGER_NAME','data']
       .apply(lambda x: x.set_index('MANAGER_NAME')['data'].to_dict())
       .to_json()
       )
```
